```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.cross_decomposition import PLSRegression as plsr
C_change_reshape=np.nan_to_num(C_change_reshape)
X=C_change_vyvanse
y=Clinical_change_vyvanse[:,1].reshape(-1,1)
y_placebo=Clinical_change_placebo[:,1].reshape(1,-1)
#pd.DataFrame(C_change_vyvanse)
```

```python
print(y.shape)
cca = plsr(n_components=3)
cca.fit(X,y)
y_predict=cca.predict(C_change_placebo)
print(y_predict.reshape(1,-1).shape)
# print(y_predict)
# print(y)
#plt.scatter(y.reshape(1,-1),y_predict.reshape(1,-1))
plt.scatter(y_placebo,y_predict.reshape(1,-1))

print(np.corrcoef(y_placebo,y_predict.reshape(1,-1)))
#print(cca.score(X,y))
z1=cca.coef_
print(z1.shape)
```
